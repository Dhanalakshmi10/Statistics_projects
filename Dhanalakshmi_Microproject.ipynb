## Hypothesis testing for Correlation

* Though we do see that up until now the age plays a major role, there seems to be some correlation between other components of the cement mixture and it's strength.

```python
#(On a side note it seems from the micro project data that is given we only need to establish the correlation)
```

Hypothesis:

* H0: There is NO correlation between Concrete Compressive Strength and Other Component (Eg: Cement) Rho=0
* H1: There is a correlation between Concrete Compressive Strength and Other Component (Eg: Cement)  Rho!=0

* Reason why we are considering this test 

```python
alpha=0.05 # that is risk rate is 5% and Confidence Level is 95%.
```

First I have used Kendal Tau as it does not assume a linear relationship or normal distribution and it is much less sensitive to outliers than Pearson's correlation.

```python
#A function to check each component with concrete strength.
def Strength_Components_Correlation_kendalltau(Component,Component_name):
    tau,p_value=st.kendalltau(df.Concrete_Strength,Component)
    
    print(end='\n')
    print(f'Component: {Component_name}')
    if p_value>alpha:
        print("Reject Null Hypothesis H0")
        print(f"There is NO correlation between Concrete Compressive Strength and {Component_name} of correlation coeff {tau}")
    else:
        print("Fail To Reject (Accept) Null Hypothesis H1")
        print(f"There is a correlation between Concrete Compressive Strength and {Component_name} of correlation coeff {tau}")
        
```

```python
# Calling the function.
Components_columns=['Age','Cement','Water','Superplasticizer','Fine_Aggrerate','Coarse_Aggregate','Fly_Ash','Blast_Furnace_Slag']
for components in Components_columns:
    Strength_Components_Correlation_kendalltau(df[components],components)
    
```

I have also considered Peasrsonr as this was the only thing taught to us in class

```python
def Strength_Components_Correlation_Pearson(Component,Component_name):
    r,p_value=st.pearsonr(df.Concrete_Strength,Component)
    print(end='\n')
    print(f'Component: {Component_name}')
    if p_value>alpha:
        print("Reject Null Hypothesis H0")
        print(f"There is NO correlation between Concrete Compressive Strength and {Component_name} of correlation coeff {r}")
    else:
        print("Fail To Reject (Accept) Null Hypothesis H1")
        print(f"There is a correlation between Concrete Compressive Strength and {Component_name} of correlation coeff {r}")
```

```python
Components_columns=['Age','Cement','Water','Superplasticizer','Fine_Aggrerate','Coarse_Aggregate','Fly_Ash','Blast_Furnace_Slag']
for components in Components_columns:
    Strength_Components_Correlation_Pearson(df[components],components)
    
```

Now that we have checked the correlation we see some correlation through Kendall/ Pearson we will visualize it.
