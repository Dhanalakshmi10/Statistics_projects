### Test of Mean

Assumptions:
#### Test of Normality:
H0:The selling price is normally distributed
H1:The selling price is not notmally distributed.

#### Test of Varaince:
H0:The sample selling price is varies from the population mean
H1:The sample selling price do not have variation from the population mean

```python
alpha=0.05 # Risk rate is 5% that is confidence level is 95%
CNG_prices=df[df.fuel == 'CNG']['selling_price']
LPG_prices=df[df.fuel == 'LPG']['selling_price']
Deisel_prices=df[df.fuel == 'Diesel']['selling_price']
Petrol_prices=df[df.fuel == 'Petrol']['selling_price']
```

```python
shapiro_results = {}
for fuel_type, prices in {'CNG': CNG_prices, 'Diesel': Deisel_prices, 'LPG': LPG_prices,'Petrol':Petrol_prices}.items():
    stat, p = st.shapiro(prices)
    shapiro_results[fuel_type] = (stat, p)
    if(p>alpha):
        print('Accept H0: The Fuel selling price is following normal distribution')
        print(f'Shapiro-Wilk test for {fuel_type}: \n Statistic: {stat}\n p-value: {p} \n')
    else:
        print('Fail to accept H0: The Fuel selling price is not following normal distribution and Procced with non-parametric test of Krushkal wallis')
        print(f'Shapiro-Wilk test for {fuel_type}: \n Statistic: {stat}\n p-value: {p} \n')
```

```python
tat,p_value=st.kruskal(CNG_prices, LPG_prices, Deisel_prices,Petrol_prices)
print("kruskal Wallis test for :",end="")
print(f'\nStatistic: {stat} \np-value: {p_value}\n')


if p_value > alpha:
    print("Accpet the null hypothesis: There is no significant difference in selling prices based on fuel type.")
else:
    print("Reject the null hypothesis: There is a significant difference in selling prices based on fuel type.")
    
```

* b)	As a data analyst in a car dealership company, you are analyzing the company's sales data to identify any significant differences in the selling prices of cars based on the fuel type.  If there is a statistically significant difference in the selling price distribution for different fuel types, which fuel types have the highest and lowest selling prices? Your task is to perform an analysis of the data and report your findings to the management.

#### Hypothesis testing for difference in selling price.

##### Test of Mean

Assumptions:
###### Test of Normality:
H0:The selling price is normally distributed
H1:The selling price is not notmally distributed.
###### Test of Varaince:
H0:The sample selling price is varies from the population mean
H1:The sample selling price do not have variation from the population mean

```python
alpha=0.05
CNG_prices=df[df.fuel == 'CNG']['selling_price']
LPG_prices=df[df.fuel == 'LPG']['selling_price']
Deisel_prices=df[df.fuel == 'Diesel']['selling_price']
Petrol_prices=df[df.fuel == 'Petrol']['selling_price']
```

```python
#Test of normality
shapiro_results = {}
for fuel_type, prices in {'CNG': CNG_prices, 'Diesel': Deisel_prices, 'LPG': LPG_prices,'Petrol':Petrol_prices}.items():
    stat, p = st.shapiro(prices)
    shapiro_results[fuel_type] = (stat, p)
    if(p>alpha):
        print('Accept H0: The Fuel selling price is following normal distribution')
        print(f'Shapiro-Wilk test for {fuel_type}: \n Statistic: {stat}\n p-value: {p} \n')
    else:
        print('Fail to accept H0: The Fuel selling price is not following normal distribution and Procced with non-parametric test of Krushkal wallis')
        print(f'Shapiro-Wilk test for {fuel_type}: \n Statistic: {stat}\n p-value: {p} \n')
```

```python
levene_results = st.levene(CNG_prices, LPG_prices, Deisel_prices,Petrol_prices)
print("Levene's test for homogeneity of variances:",end="")
print(f'\nStatistic: {levene_results.statistic} \np-value: {levene_results.pvalue}\n')

if(levene_results.pvalue>alpha):
    print('Levene is passed')
else:
    print('Levenne test is failed to accept the null hypothesis so there is variance given and population procced with Kurhskal wallis test')
```

```python
stat,p_value=st.kruskal(CNG_prices, LPG_prices, Deisel_prices,Petrol_prices)
print("kruskal Wallis test for :",end="")
print(f'\nStatistic: {stat} \np-value: {p_value}\n')


if p_value > alpha:
    print("Accpet the null hypothesis: There is no significant difference in selling prices based on fuel type.")
else:
    print("Reject the null hypothesis: There is a significant difference in selling prices based on fuel type.")
    
```
